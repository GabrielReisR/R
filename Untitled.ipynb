```python
tickers = ["ITUB4.SA", "FLRY3.SA", "WEGE3.SA", "SQIA3.SA", "EGIE3.SA", "MGLU3.SA"]

carteira = web.get_data_yahoo(tickers, period = "3months")["Adj Close"]

ibov = web.get_data_yahoo("^BVSP", period = "3months")["Adj Close"]
```

```python
sns.set()

carteira.plot(figsize = (15,5))
```

```python
carteira_normalizada = (carteira / carteira.iloc[0])*10000

ibov_normalizado = (ibov / ibov.iloc[0])*60000

carteira_normalizada["saldo"] = carteira_normalizada.sum(axis = 1)
```

```python
carteira_normalizada["saldo"].plot(figsize = (18,8), label = "Minha Carteira")
ibov_normalizado.plot(label = "Ibov")
plt.legend()
```
